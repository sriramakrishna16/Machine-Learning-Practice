### Feature Scaling and Train_Test Split

### What is Feature Scaling?
Feature Scaling is the process of bringing numerical features to a similar range so that features with larger values do not dominate the machine learning model.

### Normalization
Normalization rescales every feature to a fized range , typically 0 to 1 by using minimum and maximum values of that feature

Formula = x - xmin / xmax - xmin

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame({"salary": [200,300,400,500,600]})
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(df)

scaled_data  #it gives array format so we use pd dataframe

minmax_used = pd.DataFrame(scaled_data, columns = df.columns)
minmax_used
```

Standardization Feature or Z_score scaling

z_score is used when there are outliers in the data , usually minmax not works well as the normall range values are compressed to zero when there are outliers.

so z_score calculate the values how far from the average better at handling outliers than normalization

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()

scaled2_data = scaler2.fit_transform(df)

z_data = pd.DataFrame(scaled2_data, columns=df.columns)

z_data
```

Entire process of data preprocessing

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

here the column deck has null 688 vallues which raises the bad impact on data 

so i am dropping the dect column

```python
df = df.drop("deck",axis=1)
df
```

Now the age column has 177 null values

so i am filling null values using mean

```python
df["age"] = df["age"].fillna(df["age"].mean())
df.isnull().sum()
```

embarked has 2 null values , filling na by using forward fill

embark_town also has 2 null values , filling na by using backward fill

```python
df["embarked"] = df["embarked"].ffill()
df["embark_town"] = df["embark_town"].bfill()

df.isnull().sum()
```

Now all Null values are removed 

next step is to remove duplicated values

```python
df.duplicated().value_counts()
```

here the 111 values repeating the same data, so we have to remove this

```python
df = df.drop_duplicates()
df.duplicated().value_counts()
```

```python
df
```

now , converting categorical data into numerical data using encoding techniques

for ordered categories , Label Encoder is best 

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
oh = OneHotEncoder(sparse_output = False)

col = ["sex", "who", "adult_male", "alive", "alone"]

for i in col:
    df[i] = le.fit_transform(df[i])

df
```

```python
one_hot = oh.fit_transform(df[["embark_town", "class" , "embarked"]])

one_hot = pd.DataFrame(one_hot, columns = oh.get_feature_names_out(),index=df.index)

one_hot
```

```python
df_new = pd.concat([df, one_hot] , axis=1)
df_new
```

now remove the categorical columns

```python
df_new = df_new.drop("embarked", axis=1)
df_new = df_new.drop("embark_town", axis=1)
df_new = df_new.drop("class", axis=1)
```
